```python
from unsloth import FastLanguageModel
import torch

# --- CẤU HÌNH ĐƯỜNG DẪN ---
model_path = "/content/drive/MyDrive/checkpoint-5625"
max_seq_length = 2048

print(f"Đang load model từ: {model_path}")

# Load model
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = model_path,
    max_seq_length = max_seq_length,
    dtype = None,
    load_in_4bit = True,
)


FastLanguageModel.for_inference(model)
```

```python
def chat_voi_bot(question):
    print(f"\nUser: {question}")

    messages = [
        {"role": "system", "content": "Bạn là trợ lý am hiểu lịch sử Việt Nam."},
        {"role": "user", "content": question}
    ]

    input_ids = tokenizer.apply_chat_template(
        messages,
        add_generation_prompt=True,
        return_tensors="pt"
    ).to("cuda")

    outputs = model.generate(
        input_ids,
        max_new_tokens=512,
        temperature=0.7,
        use_cache=True
    )

    response = tokenizer.batch_decode(outputs)
    clean_response = response[0].split('<|im_start|>assistant')[-1].replace('<|im_end|>', '').strip()

    print(f"Bot: {clean_response}")
    print("-" * 50)


chat_voi_bot("Chiến thắng Điện Biên Phủ có ý nghĩa như thế nào?")
chat_voi_bot("Tại sao nhà Trần lại 3 lần thắng quân Nguyên Mông?")
chat_voi_bot("Ai là vị vua cuối cùng của triều đại phong kiến Việt Nam?")
```

```python
from unsloth import FastLanguageModel, PatchDPOTrainer
from unsloth import is_bfloat16_supported
from trl import DPOTrainer, DPOConfig
from datasets import load_dataset


PatchDPOTrainer()
max_seq_length = 2048

# --- ĐƯỜNG DẪN  ---
model_sft_path = "/content/drive/MyDrive/checkpoint-5625"
data_dpo_path = "/content/drive/MyDrive/My_Chatbot_Data/dpo_train.json"
output_dir_dpo = "/content/drive/MyDrive/DPO_Final_Model"
# -----------------------------------

print(f"Đang load model SFT từ: {model_sft_path}")

# 2. Load Model SFT
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = model_sft_path,
    max_seq_length = max_seq_length,
    dtype = None,
    load_in_4bit = True,
)

# 3. Load Data DPO
dataset = load_dataset("json", data_files=data_dpo_path, split="train")

# 4. Cấu hình Train DPO

dpo_trainer = DPOTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset,
    max_length = max_seq_length,
    max_prompt_length = 1024,
    beta = 0.1,


    args = DPOConfig(
        per_device_train_batch_size = 1,
        gradient_accumulation_steps = 8,
        warmup_steps = 10,
        num_train_epochs = 1,
        learning_rate = 5e-6,
        fp16 = not is_bfloat16_supported(),
        bf16 = is_bfloat16_supported(),
        logging_steps = 10,
        optim = "adamw_8bit",
        weight_decay = 0.01,
        lr_scheduler_type = "linear",
        seed = 3407,
        output_dir = output_dir_dpo,
        report_to = "none",
        # Các tham số riêng của DPOConfig
        rpo_alpha = None,
        model_init_kwargs = None
    ),
)

print("BẮT ĐẦU HUẤN LUYỆN DPO...")
dpo_trainer.train()

# 5. Lưu model cuối cùng
print("Đang lưu model DPO Final...")
dpo_trainer.save_model(output_dir_dpo)
tokenizer.save_pretrained(output_dir_dpo)
print(f"XONG! Model DPO xịn nhất đã nằm tại: {output_dir_dpo}")
```
